```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (
    classification_report, accuracy_score
)
from scipy.stats import mode
```

```python
excel_file = pd.ExcelFile("Mine_Dataset_edited.xlsx")
print("Available Sheets:", excel_file.sheet_names)

mine_df = pd.read_excel(excel_file, sheet_name="Mine_Data")
```

```python
print("\nMean:\n", mine_df.mean(numeric_only=True))
print("\nMedian:\n", mine_df.median(numeric_only=True))
print("\nMode:\n", mine_df.mode(numeric_only=True).iloc[0])
```

```python
print("\nMissing Values:\n", mine_df.isnull().sum())
mine_df.fillna(mine_df.mean(numeric_only=True), inplace=True) ##updates the original DataFrame (no need to assign it to a new variable).
print("Total Missing After Fill:", mine_df.isnull().sum().sum())
```

```python
columns_to_check = ['V', 'H', 'M', 'S']

for col in columns_to_check:
    col_data = mine_df[col]
    mean_val = np.mean(col_data)
    median_val = np.median(col_data)
    mode_val = mode(col_data, keepdims=True).mode[0] #Keeps output format consistent for safety
    skewness = 3 * (mean_val - median_val)
    skew_type = "Symmetrical" if np.isclose(skewness, 0, atol=1e-3) else (
        "Positively Skewed" if skewness > 0 else "Negatively Skewed")

    print(f"\nColumn: {col}")
    print(f"Mean: {mean_val:.4f}, Median: {median_val:.4f}, Mode: {mode_val:.4f}")
    print(f"Skewness: {skewness:.4f} → {skew_type}")

    plt.figure(figsize=(6, 4))
    sns.kdeplot(col_data, fill=True, color='skyblue', linewidth=2) #plots a smooth curve of the column’s distribution.
    plt.axvline(mode_val, color='red', linestyle='--', label='Mode') #Adds vertical lines
    plt.axvline(median_val, color='green', linestyle='--', label='Median')
    plt.axvline(mean_val, color='blue', linestyle='--', label='Mean')
    plt.title(f"{col} - {skew_type}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
Q1 = mine_df.quantile(0.25)
Q3 = mine_df.quantile(0.75)
IQR = Q3 - Q1


print("Q1 (25th percentile):")
print(Q1)
print("\nQ3 (75th percentile):")
print(Q3)
print("\nIQR (Q3 - Q1):")
print(IQR)

outliers = mine_df[((mine_df < (Q1 - 1.5 * IQR)) | (mine_df > (Q3 + 1.5 * IQR))).any(axis=1)] # keeps rows where any column contains an outlier.
print("\nOutliers Detected:")
print(outliers)

mine_df_clean = mine_df[~((mine_df < (Q1 - 1.5 * IQR)) | (mine_df > (Q3 + 1.5 * IQR))).any(axis=1)]

mine_df_clean = mine_df_clean.fillna(mine_df_clean.mean(numeric_only=True))

print("\nCleaned Dataset (no outliers, missing values filled):")
print(mine_df_clean)
print(f"\n✅ Data shape after removing outliers: {mine_df_clean.shape}")
```

```python
# KDE Plots (After Outlier Removal)
for col in columns_to_check:
    col_data = mine_df_clean[col]
    mean_val = np.mean(col_data)
    median_val = np.median(col_data)
    mode_val = mode(col_data, keepdims=True).mode[0]
    skewness = 3 * (mean_val - median_val)
    skew_type = "Symmetrical" if np.isclose(skewness, 0, atol=1e-3) else (
        "Positively Skewed" if skewness > 0 else "Negatively Skewed")

    print(f"\n[CLEANED] Column: {col}")
    print(f"Mean: {mean_val:.4f}, Median: {median_val:.4f}, Mode: {mode_val:.4f}")
    print(f"Skewness: {skewness:.4f} → {skew_type}")

    plt.figure(figsize=(6, 4))
    sns.kdeplot(col_data, fill=True, color='orange', linewidth=2)
    plt.axvline(mode_val, color='red', linestyle='--', label='Mode')
    plt.axvline(median_val, color='green', linestyle='--', label='Median')
    plt.axvline(mean_val, color='blue', linestyle='--', label='Mean')
    plt.title(f"[Cleaned] {col} - {skew_type}")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
features = ['V', 'H', 'S']
X = mine_df_clean[features]
y = mine_df_clean['M']

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
    #ensures both training and test sets have the same class distribution (useful if your data is imbalanced).
)

print(f"\n📚 Training Set: {X_train.shape}")
print(f"🧪 Testing Set: {X_test.shape}")
```

```python
# ========== 7. Classification Model ==========
clf_model = RandomForestClassifier(random_state=42)
clf_model.fit(X_train, y_train)
y_pred_class = clf_model.predict(X_test)

print("\n📊 Random Forest Classification Report:")
print(f"Predicted Classes:\n{y_pred_class}\n")
print(classification_report(y_test, y_pred_class))
print("✅ Accuracy:", accuracy_score(y_test, y_pred_class))
```

```python
# Adding K-Nearest Neighbors (KNN)
from sklearn.neighbors import KNeighborsClassifier

# K-Nearest Neighbors (KNN)
knn_model = KNeighborsClassifier()
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

print("\n📊 KNN Classification Report:")
print(f"Predicted Classes:\n{y_pred_knn}\n")
print(classification_report(y_test, y_pred_knn))
print("✅ KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
```

```python
# Adding Logistic Regression
from sklearn.linear_model import LogisticRegression

# Logistic Regression
logreg_model = LogisticRegression(random_state=42)
logreg_model.fit(X_train, y_train)
y_pred_logreg = logreg_model.predict(X_test)

print("\n📊 Logistic Regression Classification Report:")
print(f"Predicted Classes:\n{y_pred_logreg}\n")
print(classification_report(y_test, y_pred_logreg))
print("✅ Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_logreg))
```

```python
# Adding Support Vector Machine (SVM)
from sklearn.svm import SVC

# Initialize and train the model
# Support Vector Machine (SVM)
svm_model = SVC(random_state=42)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print("\n📊 SVM Classification Report:")
print(f"Predicted Classes:\n{y_pred_svm}\n")
print(classification_report(y_test, y_pred_svm))
print("✅ SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
```

```python
# Adding Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier

# Gradient Boosting Classifier
gbc_model = GradientBoostingClassifier(random_state=42)
gbc_model.fit(X_train, y_train)
y_pred_gbc = gbc_model.predict(X_test)

print("\n📊 Gradient Boosting Classification Report:")
print(f"Predicted Classes:\n{y_pred_gbc}\n")
print(classification_report(y_test, y_pred_gbc))
print("✅ Gradient Boosting Accuracy:", accuracy_score(y_test, y_pred_gbc))
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.linear_model import LogisticRegression

# Base models
base_models = [
    ('rf', RandomForestClassifier(random_state=42)),
    ('gb', GradientBoostingClassifier(random_state=42))
]

# Final model (meta-model)
meta_model = LogisticRegression()

# Stacking ensemble
stack_model = StackingClassifier(estimators=base_models, final_estimator=meta_model)

# Train
stack_model.fit(X_train, y_train)

# Predict
y_pred_stack = stack_model.predict(X_test)

# Evaluate
print("\n📊 Stacked Model Classification Report:")
print(classification_report(y_test, y_pred_stack))
print("✅ Stacked Model Accuracy:", accuracy_score(y_test, y_pred_stack))
```

```python
#Confusion Matrix & Classification Report

from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

models = {
    "Random Forest": clf_model,
    "K-Nearest Neighbors": knn_model,
    "Logistic Regression": logreg_model,
    "Support Vector Machine": svm_model,
    "Gradient Boosting": gbc_model,
    "Stacked Model" : stack_model
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'{name} - Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    print("\n")
```

```python
#10-Fold Cross-Validation Code

from sklearn.model_selection import cross_val_score, StratifiedKFold

# Use the same model names you defined earlier
models = {
    "Random Forest": clf_model,
    "K-Nearest Neighbors": knn_model,
    "Logistic Regression": logreg_model,
    "Support Vector Machine": svm_model,
     "Gradient Boosting": gbc_model,
    "Stacked Model" : stack_model
}

cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

for name, model in models.items():
    scores = cross_val_score(model, X_scaled, y, cv=cv, scoring='accuracy')
    print(f"{name} - Cross-Validated Accuracy: {scores.mean():.4f} ± {scores.std():.4f}")
```

```python
#ROC Curves and AUC Code

from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_auc_score, roc_curve, auc
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import LabelBinarizer

# Binarize the output (needed for multiclass ROC)
lb = LabelBinarizer()
y_test_binarized = lb.fit_transform(y_test)

# If 2-class, label_binarize returns (n_samples, ), so we reshape
if y_test_binarized.ndim == 1:
    y_test_binarized = y_test_binarized.reshape(-1, 1)

plt.figure(figsize=(10, 7))

for name, model in models.items():
    try:
        classifier = OneVsRestClassifier(model)
        classifier.fit(X_train, y_train)

        if hasattr(classifier, "predict_proba"):
            y_score = classifier.predict_proba(X_test)
        else:
            y_score = classifier.decision_function(X_test)

        # Compute ROC curve and ROC area for each class
        fpr = dict()
        tpr = dict()
        roc_auc = dict()
        n_classes = y_test_binarized.shape[1]

        for i in range(n_classes):
            fpr[i], tpr[i], _ = roc_curve(y_test_binarized[:, i], y_score[:, i])
            roc_auc[i] = auc(fpr[i], tpr[i])

        # Compute micro-average ROC curve and ROC area
        all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
        mean_tpr = np.zeros_like(all_fpr)

        for i in range(n_classes):
            mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])

        mean_tpr /= n_classes
        plt.plot(all_fpr, mean_tpr, label=f'{name} (avg. AUC = {np.mean(list(roc_auc.values())):.2f})')

    except Exception as e:
        print(f"⚠️ Skipping {name} for ROC: {e}")

plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.title("Multiclass ROC Curve Comparison")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
